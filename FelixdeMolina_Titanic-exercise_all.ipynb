9. What is the name of the oldest person who died in the Titanic? Was he or she travelling alone or had any family travelling with them?  



```python
# Oldest person who died in the Titanic
oldest_died = titanic[titanic['survived'] == 0].sort_values('age', ascending = False)
print(oldest_died.iloc[0])
```

In order to give an answer to the second question you should find out which columns give you that info. Usually part of your job as a Data Scientist will be get to know the dataset which you are working with. In this case the columns which give you that info are the following: 
        - 'sibsp' Number of Siblings/Spouses Aboard
        - 'parch' Number of Parents/Children Aboard

```python
# Was he or she travelling alone or had any family travelling with?
'''DespuÃ©s de media hora probando cosas se me ocurre generar una columna en una copia del DataFrame original'''
titanic_alone = titanic.copy()
oldest_person_died = titanic_alone[titanic_alone['survived'] == 0].sort_values('age', ascending = False)
titanic_alone['traveling_alone'] = (titanic['sibsp'] == 0) & (titanic['parch'] == 0)
if oldest_person_died['traveling_alone'].iloc[0]:
    print(f"The oldest person who died was traveling alone in the Titanic")
else:
    print(f"The oldest person who died was not traveling alone in the Titanic")
```

```python
'''Como extra, voy a colocar la columna traveling_alone en la nueva tabla a continuaciÃ³n de sibsp y parch'''
traveling_alone = titanic_alone.pop('traveling_alone') #Este mÃ©todo no borra definitivamente una columna, sino que lo almacena en una variable del mismo nombre para que no se pierda el contenido.
titanic_alone.insert(7, 'traveling_alone', traveling_alone)
titanic_alone
```

34. Using [`kdeplot`]("https://seaborn.pydata.org/generated/seaborn.kdeplot.html") from `seaborn` represent those who not survived distributed by age.<>

Hint: First you should "filter" the `titanic` dataset where the column "survived" is 0, indexing the column `"age"` only.

Arguments you should pass to the function: 

    - color = "red"
    - label = "Not Survived"
    - shade = True
    
You should get something like this: 

![kdeplot](./img/kdeplotsur.png)

```python
not_survived = titanic[titanic['survived'] == 0]

sns.kdeplot(data=not_survived['age'], color = 'red', label='Not Survived', fill=True)

plt.title('Age distribution of people who did not survive the Titanic')
plt.xlabel('Age')
plt.ylabel('Not Survived Density')

plt.show()
```
